```python
#Importe de librerias
from selenium import webdriver



from selenium.webdriver.chrome.service import Service

from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.by import By


import pandas as pd
import os
import openpyxl
import collections
from selenium.webdriver.common import keys

```

```python


######## Abriendo la pagina
url = "https://www.linkedin.com/jobs/search?keywords=Analista%20De%20Datos&location=Panam%C3%A1%2C%20Panam%C3%A1%2C%20Panam%C3%A1&geoId=100508521&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0"
driver = webdriver.Chrome((ChromeDriverManager().install()))
driver.get(url)
driver.maximize_window()
time.sleep(3)

##### Definiendo primeras variables

company_name = []
job_title = []
link_href = []
posting_time = []
applicants_num = []
contenido_desc = []
nivel_antiguedad = []
tipo_empleo = []
funcion_laboral = []
sector_ind = []
##### Empezando a scrape
links = driver.find_elements(By.CLASS_NAME, 'base-card__full-link')
for link in links:
    try:
        time.sleep(2)
        driver.execute_script("arguments[0].click();", link)
        time.sleep(2)
        ##### Seccion con la data

        infocard = driver.find_elements(By.XPATH, '/html/body/div[1]/div/section')
        for info in infocard:
            ##### Esperando a que cargue la seccion con la data y extrayendo la info

            time.sleep(2)
            job = info.find_element(By.XPATH, '/html/body/div[1]/div/section/div[2]/section/div/div[1]/div/a/h2')
            company = info.find_element(By.XPATH,
                                        '/html/body/div[1]/div/section/div[2]/section/div/div[1]/div/h4/div[1]/span[1]')
            postingtime = info.find_element(By.XPATH,
                                            '/html/body/div[1]/div/section/div[2]/section/div/div[1]/div/h4/div[2]/span[1]')
            applicants = info.find_element(By.XPATH,
                                           '/html/body/div[1]/div/section/div[2]/section/div/div[1]/div/h4/div[2]/span[2]')
            showmore = info.find_element(By.XPATH,
                                         '/html/body/div[1]/div/section/div[2]/div/section[1]/div/div/section/button[1]')

            ##### Guardando la info en el dict()

            company_name.append(company.text)
            job_title.append(job.text)
            link_href.append(link.get_attribute('href'))
            posting_time.append(postingtime.text)
            applicants_num.append(applicants.text)

            try:
                ##### Clickeando en show more y repitiendo el proceso para guardar lo que falta

                driver.execute_script("arguments[0].click();", showmore)
                time.sleep(2)
                contenido = info.find_element(By.XPATH,
                                              '/html/body/div[1]/div/section/div[2]/div/section[1]/div/div/section/div/ul[1]')
                antiguedad = info.find_element(By.XPATH,
                                               '/html/body/div[1]/div/section/div[2]/div/section[1]/div/ul/li[1]/span')
                tipodeempleo = info.find_element(By.XPATH,
                                                 '/html/body/div[1]/div/section/div[2]/div/section[1]/div/ul/li[2]/span')
                funcionlaboral = info.find_element(By.XPATH,
                                                   '/html/body/div[1]/div/section/div[2]/div/section[1]/div/ul/li[3]/span')
                sector = info.find_element(By.XPATH,
                                           '/html/body/div[1]/div/section/div[2]/div/section[1]/div/ul/li[4]/span')

                contenido_desc.append(contenido.text)
                nivel_antiguedad.append(antiguedad.text)
                tipo_empleo.append(tipodeempleo.text)
                funcion_laboral.append(funcionlaboral.text)
                sector_ind.append(sector.text)

            except:
                pass
    except:
        pass
```

```python
table_dict = {'Compania': company_name, 'Posicion': job_title, 'Link web': link_href, 
              'Fecha de publicacion': posting_time, 'Numbero de aplicantes': applicants_num, 
              'Descripcion del puesto': contenido_desc, 'Experiencia requerida': nivel_antiguedad,
              'Tipo de empleo': tipo_empleo, 'Funcion laboral': funcion_laboral, 'Sector': sector_ind}
```

```python
DF = pd.DataFrame.from_dict(table_dict, orient='index')
```

```python
newdict = pd.DataFrame.to_dict(DF)
FinalDF = pd.DataFrame.from_dict(newdict, orient='index')
FinalDF
```
